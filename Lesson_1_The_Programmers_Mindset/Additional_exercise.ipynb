### **🛠️ Lesson 1: The Programmer's Mindset**
**Exercise 1: The Hardware Match-up**
Imagine you are a specialized worker in a large factory (the computer). Match the following computer components to their real-world human equivalents based on their roles:


* **CPU**


* **Main Memory (RAM)**


* **Secondary Memory**


* **Input Devices**

**Match with:**


    A. A massive library of filing cabinets where books stay forever until they are checked out or thrown away.


    B. A hyper-focused worker who asks "What would you like me to do next?" billions of times a second.


    C. The worker's eyes and ears, allowing them to perceive what is happening outside the factory.


    D. A small, fast chalkboard where the worker scribbles notes that are erased the moment the lights go out.

**Exercise 2: Identifying the "Story" Pattern**
Every program follows a specific flow. Identify which pattern **(Sequential, Conditional, or Repeated)** is being described in each scenario:

* **Scenario A:** You ask the user for their age. If they are over 18, you print "Welcome"; otherwise, you print "Access Denied".


* **Scenario B:** You take a list of 1,000 students and calculate the average grade for each one, one by one, until the list is finished.


* **Scenario C:** You prompt for a name, save it to a variable, and then print a "Hello" message using that name .

### Solutions:
**Exercise 1: The Hardware Match-up**

* **CPU** - B. A hyper-focused worker who asks "What would you like me to do next?" billions of times a second.

* **Main Memory (RAM)** - D. A small, fast chalkboard where the worker scribbles notes that are erased the moment the lights go out.

* **Secondary Memory** - A. A massive library of filing cabinets where books stay forever until they are checked out or thrown away.

* **Input Devices** - C. The worker's eyes and ears, allowing them to perceive what is happening outside the factory.

**Exercise 2: Identifying the "Story" Pattern**

* **Scenario A:** Conditional - The program makes a decision based on the user's input (age).

* **Scenario B:** Repeated - The program performs the same action (calculating average grade) for each student in the list.

* **Scenario C:** Sequential - The program executes steps in a specific order: prompting for a name, saving it, and then printing a message.
