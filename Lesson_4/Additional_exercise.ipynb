**Requirements:**

* Inside the function, calculate the Fahrenheit equivalent using the formula: `(c_temp * 9/5) + 32`.
* The function must `return` the calculated Fahrenheit value (it should NOT use `print()` inside the function).
* Outside the function, prompt the user for a Celsius temperature, convert it to a float, call your function, store the result in a variable, and then print the result.

### Exercise 4: The Event Ticket Kiosk
**Goal:** Combine Chapter 3 logic inside a Chapter 4 function to create a complete mini-application.

**Requirements:**
* Define a function called `get_ticket_price` that takes one parameter: `age` (an integer).
* Inside the function, use `if/elif/else` to determine the price:

  * Age 12 and under: $10

  * Age 13 to 64: $20

  * Age 65 and older: $15

* `return` the price.

* Write a main script that asks the user for their age, uses a `try/except` block to ensure they typed a valid integer, calls `get_ticket_price(age)`, and prints: *"Your ticket costs: $[price]"*.

```python
# solution for Exercise 3

# define the function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(c_temp):
    fahrenheit = (c_temp * 9/5) + 32
    return fahrenheit

# prompt the user for a Celsius temperature
celsius_input = input("Please enter a temperature in Celsius: ")

try:
    # try to convert the input to a float
    celsius_temp = float(celsius_input)

    # call the function and store the result
    fahrenheit_temp = celsius_to_fahrenheit(celsius_temp)

    # print the result
    print(f"{celsius_temp} degrees Celsius is equal to {fahrenheit_temp} degrees Fahrenheit.")
except ValueError:
    # handle the case where the input is not a valid number
    print("Error: Please enter a valid number")
```

```python
# solution for Exercise 4

# define the function to get ticket price based on age
def get_ticket_price(age):
    if age <= 12:
        return 10
    elif 13 <= age <= 64:
        return 20
    else:
        return 15

# main script
age_input = input("Please enter your age: ")
try:
    # try to convert the input to an integer
    age = int(age_input)

    # call the function to get the ticket price
    price = get_ticket_price(age)

    # print the ticket price
    print(f"Your ticket costs: ${price}")
except ValueError:
    # handle the case where the input is not a valid integer
    print("Error: Please enter a valid integer for age")
```
