**Exercise 1: The Expression Master**

In your Master's studies, you will often deal with complex data expressions. For each Python expression below, determine the Final Value and the Data Type (e.g., int, float, or str):

* x = 15 + 5 * 2

* y = 10 / 2

* z = "Python" + "101"

* w = int(98.6) + 1

```python
# Solutions:

x = 15 + 5 * 2
print(f"x: Final Value = {x}, Data Type = {type(x)}")

y = 10 / 2
print(f"y: Final Value = {y}, Data Type = {type(y)}")

z = "Python" + "101"
print(f"z: Final Value = {z}, Data Type = {type(z)}")

w = int(98.6) + 1
print(f"w: Final Value = {w}, Data Type = {type(w)}")
```
