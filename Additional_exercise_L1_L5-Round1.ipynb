### 🎯 **Chapter 1 Boss Exercise — "The Smart Greeter"**

**What it tests:** Sequential flow, conditional flow, repeated flow, scripts, debugging mindset.

Write a complete Python script (`.py` file) that does the following:

1. Ask the user for their name
2. Ask the user for their age
3. If age is under 18, print: `"Welcome [name], you are a junior programmer!"`
4. If age is 18 to 64, print: `"Welcome [name], you are a professional programmer!"`
5. If age is 65 or above, print: `"Welcome [name], you are a wise programmer!"`
6. After the greeting, use a loop to print the user's name 3 times, each time on a new line with its number, like:

1 - Alice

2 - Alice

3 - Alice


---

### Constraints:

- You must use all 3 program flows (sequential, conditional, repeated)
- The loop must use a variable that changes each iteration
- No hardcoded names — everything comes from `input()`

````python
# Smart Greeter Script
# Step 1: Ask for user's name
name = input("Please enter your name: ")
# Step 2: Ask for user's age
age = int(input("Please enter your age: "))
# Step 3: Conditional greetings based on age
if age < 18:
    print(f"Welcome {name}, you are a junior programmer!")
elif 18 <= age < 65:
    print(f"Welcome {name}, you are a professional programmer!")
else:
    print(f"Welcome {name}, you are a wise programmer!")

# Step 4: Loop to print the user's name 3 times
for i in range(3):
    print(f"{i+1} - {name}")

# step 4 method 2:
n = 0
while n < 3:
    n += 1
    print(f"{n} - {name}")
````

### 🎯 **Chapter 2 Boss Exercise — "The Pay Calculator"**

**What it tests:** Variables, operators, type conversion, input, operator precedence, comments.

Write a program that:

1. Asks the user for **hours worked** and **hourly rate**
2. Calculates **gross pay** (hours × rate)
3. Calculates a **10% tax** on the gross pay
4. Calculates **net pay** (gross pay minus tax)
5. Prints a clean summary like:

```python
----- Pay Summary -----
Hours: 40
Rate: 15.00
Gross Pay: 600.00
Tax (10%): 60.00
Net Pay: 540.00
```

---

### Constraints:

- All inputs must be properly converted from strings to numbers
- You must use at least **5 different variables** with mnemonic names
- Add a comment above each calculation block explaining what it does
- The result of `/` must be a float — do not use `//` here
- All printed numbers must come from variables, not recalculated inline

````python
# Pay Calculator Script
# Step 1: Get user input for hours worked and hourly rate
hours_worked = float(input("Enter hours worked: "))
hourly_rate = float(input("Enter hourly rate: "))
# Step 2: Calculate gross pay
gross_pay = hours_worked * hourly_rate
# Step 3: Calculate tax (10% of gross pay)
tax = gross_pay * 0.10
# Step 4: Calculate net pay (gross pay minus tax)
net_pay = gross_pay - tax
# Step 5: Print the pay summary
print("----- Pay Summary -----")
print(f"Hours: {hours_worked}")
print(f"Rate: {hourly_rate:.2f}")
print(f"Gross Pay: {gross_pay:.2f}")
print(f"Tax (10%): {tax:.2f}")
print(f"Net Pay: {net_pay:.2f}")
````

### 🎯 **Chapter 3 Boss Exercise — "The Grade Classifier"**

**What it tests:** Boolean expressions, all conditional structures, try/except, logical operators.

Write a program that:

1. Asks the user to enter a **score (0–100)**
2. Uses `try/except` to catch non-numeric input — if the input is not a number, print:
   `"Error: please enter a numeric score."` and stop
3. If the number is outside **0–100**, print:
   `"Error: score must be between 0 and 100."` and stop
4. Otherwise classify the score using all these grades:

| Score  | Grade         |
|--------|--------------|
| 90–100 | A — Excellent |
| 80–89  | B — Good      |
| 70–79  | C — Average   |
| 60–69  | D — Passing   |
| 0–59   | F — Failing   |

5. Also print whether the student **passed or failed** using a logical operator:

```
Score : 75
Grade : C — Average
Result: Passed
```

---

### Constraints:

- Use `try/except` for the conversion
- Use `if / elif / else` for the grade — **no nested ifs allowed**
- The pass/fail line must use a single `if/else`, and the condition must use `>=`
- The score validation (0–100 range) must use the `and` operator in one condition


````python
# Grade Classifier Script
# Step 1: Get user input for score
score_input = input("Enter a score (0-100): ")
# Step 2: Try to convert input to a number
try:
    score = float(score_input)
    # Step 3: Validate score range
    if not (score >= 0 and score <= 100):
        print("Error: score must be between 0 and 100.")
    else:
        # Step 4: Classify the score into grades
        if score >= 90:
            grade = "A — Excellent"
        elif score >= 80:
            grade = "B — Good"
        elif score >= 70:
            grade = "C — Average"
        elif score >= 60:
            grade = "D — Passing"
        else:
            grade = "F — Failing"
        # Step 5: Determine pass/fail result
        result = "Passed" if score >= 60 else "Failed"
        # Print the results
        print(f"Score : {score:.0f}")
        print(f"Grade : {grade}")
        print(f"Result: {result}")

except ValueError:
    print("Error: please enter a numeric score.")
````

### 🎯 **Chapter 4 Boss Exercise — "The Function Toolkit"**

**What it tests:** Defining functions, parameters, arguments, return values, fruitful vs void, the math module.

Build a small toolkit of **4 functions**, then use all of them in a main program:

---

### Function 1 — `greet(name, lang)`
Returns a greeting string. Support:
- `'ar'` (Arabic → "سلام")
- `'fr'` (French → "Bonjour")
- anything else → "Hello"

⚠️ It must **return** the string, not print it.

---

### Function 2 — `compute_bmi(weight_kg, height_m)`
Returns the BMI as a float using the formula: `weight / (height ** 2)` Round to **2 decimal places** using `round()`.

---

### Function 3 — `classify_bmi(bmi)`
Takes the BMI value and returns a category string:

- Under 18.5 → `"Underweight"`
- 18.5 to 24.9 → `"Normal"`
- 25 to 29.9 → `"Overweight"`
- 30 and above → `"Obese"`

---

### Function 4 — `print_report(name, lang, bmi, category)`
A **void function** (no return). Prints the full report:

```
----- Health Report -----
Bonjour, Claire!
BMI : 22.86
Category : Normal
```

---

### Main program

- Ask the user for:
  - name
  - language
  - weight
  - height
- Call all **4 functions** in the correct order
- Pass results between functions properly

---

### Constraints:

- Functions **1, 2, and 3 must use `return`** (they are fruitful)
- Function 4 must **NOT return anything** (it is void)
- The main program must store each function's result in a variable before passing it
- Use `import math` and replace `height ** 2` with `math.pow(height, 2)` in function 2





````python
import math

# Function 1: greet(name, lang) — fruitful, returns a string
def greet(name, lang):
    if lang == 'ar':
        return f"مرحبا, {name}!"
    elif lang == 'fr':
        return f"Bonjour, {name}!"
    else:
        return f"Hello, {name}!"

# Function 2: compute_bmi(weight_kg, height_m) — fruitful, returns a float
def compute_bmi(weight_kg, height_m):
    bmi = weight_kg / math.pow(height_m, 2)
    return round(bmi, 2)

# Function 3: classify_bmi(bmi) — fruitful, returns a string
def classify_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif bmi < 25:
        return "Normal"
    elif bmi < 30:
        return "Overweight"
    else:
        return "Obese"

# Function 4: print_report(name, lang, bmi, category) — VOID, no return
def print_report(name, lang, bmi, category):
    greeting = greet(name, lang)        # greet() is called INSIDE the void function
    print("----- Health Report -----")
    print(greeting)
    print(f"BMI      : {bmi}")
    print(f"Category : {category}")

# ── Main program ──────────────────────────────────────────────────────────────
name     = input("Enter your name: ")
lang     = input("Enter your language (ar/fr/other): ")
weight   = float(input("Enter your weight in kg (e.g. 70): "))
height   = float(input("Enter your height in meters (e.g. 1.75): "))  # ← meters, not cm!

# Each result stored in a variable before being passed forward
bmi      = compute_bmi(weight, height)
category = classify_bmi(bmi)
print_report(name, lang, bmi, category)   # greet() called internally here
````

🎯 **Chapter 5 Boss Exercise — "The Data Analyzer"**

**What it tests:** while loop, for loop, break, continue, all 5 loop idioms (count, sum, average, filter, boolean search), None pattern, is operator.

---

## 🧩 Write a program in two parts:

---

## 🔹 Part 1 — Data Collection (while loop)

- Keep asking the user: `"Enter a number (or 'done' to finish): "`

- Use `continue` to skip any input that starts with `#` (treat it as a comment)

- Use `break` when the user types `done`

- Store each valid number in a list called `numbers`

- Use `try/except` to reject non-numeric entries with:

Skipping invalid input.


---

## 🔹 Part 2 — Data Analysis (for loops)

Once the user is done entering data, run **all 5 analyses** on the `numbers` list and print a report:

```
-------- Analysis Report --------
Count : 6
Sum : 154
Average : 25.67
Large numbers (> 20): 41, 74
Search for 3: Found
Smallest : 3
Largest : 74
```


---

## ⚠️ Constraints:

- Part 1 must use a `while True` loop with `break` and `continue`
- Part 2 must use **separate `for` loops** for each analysis (no combining them all in one loop)
- You must implement:
  - Count
  - Sum
  - Average
  - Filter (numbers > 20)
  - Boolean search (e.g., check if 3 exists)
- Smallest and largest must use the **None pattern** (no `min()` or `max()`)
- Use the `is` operator when checking the None pattern

````python
# Data Analyzer Script
# Part 1: Data Collection
numbers = []
while True:
    user_input = input("Enter a number (or 'done' to finish): ")
    if user_input.lower() == 'done':
        break
    if user_input.startswith('#'):
        continue
    try:
        number = float(user_input)
        numbers.append(number)
    except ValueError:
        print("Skipping invalid input.")

# Part 2: Data Analysis
# Count
count = 0
for num in numbers:
    count += 1
# Sum
total_sum = 0
for num in numbers:
    total_sum += num
# Average
average = total_sum / count if count > 0 else 0
# Filter (numbers > 20)
large_numbers = []
for num in numbers:
    if num > 20:
        large_numbers.append(num)
# Boolean search for 3
found_three = False
for num in numbers:
    if num == 3:
        found_three = True
        break
# Smallest and Largest using None pattern
smallest = None
largest = None
for num in numbers:
    if smallest is None or num < smallest:
        smallest = num
    if largest is None or num > largest:
        largest = num

# Print the analysis report
if count == 0:
    print("No data entered.")
else:
    print("-------- Analysis Report --------")
    print(f"Count : {count}")
    print(f"Sum : {total_sum}")
    print(f"Average : {average:.2f}")
    print(f"Large numbers (> 20): {', '.join(str(n) for n in large_numbers) or 'None'}")
    print(f"Search for 3: {'Found' if found_three else 'Not Found'}")
    print(f"Smallest : {smallest}")
    print(f"Largest : {largest}")

````
