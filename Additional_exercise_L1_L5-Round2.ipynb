## 🎯 Chapter 5 Boss Exercise — "The Student Grade Tracker"

**What it tests:** while loop, for loop, break, continue, all 5 loop idioms, None pattern, is operator, filtering.

Write a program in two parts:

**Part 1 — Data collection (while loop):**
- Keep asking: `"Enter student score (or 'done' to finish): "`
- Use `continue` to skip any input that starts with `#`
- Use `break` when the user types `done`
- Use `try/except` to reject non-numeric input with `"Invalid score, skipping."`
- After conversion, use `continue` to skip scores outside 0–100 with `"Score out of range, skipping."`
- Store valid scores in a list called `scores`

**Part 2 — Analysis (separate for loops):**

Run all 5 analyses and print this report:
```
-------- Grade Report --------
Total students : 6
Sum of scores  : 414
Class average  : 69.0
Passed (>= 60) : 82, 75, 91, 63
Failed  (< 60) : 45, 58
Searched 100   : Not found
Highest score  : 91
Lowest score   : 45
```

### Constraints:
- Part 1 must use `while True` with `break` and `continue` (continue used **twice** — for `#` and for out-of-range)
- Part 2 must use **completely separate `for` loops** for each analysis
- Highest and lowest must use the `None` + `is` pattern — no `max()` or `min()`
- Boolean search must look for the score `100` and start with `found = False`
- Average must guard against empty list (no division by zero)
- Passed and failed lists must be printed as comma-separated values on one line each

````python
# exercise 1
secret = 7

name = input("What is your name? ")
print(f"Hello {name}! I'm thinking of a number between 1 and 10.")

won = False
for attempt in range(1, 4):
    guess = int(input(f"Attempt {attempt} — Enter your guess: "))
    if guess == secret:
        print(f"🎉 Correct {name}! You got it on attempt {attempt}!")
        won = True
        break
    elif guess < secret:
        print("Too low, try again!")
    else:
        print("Too high, try again!")

if not won:
    print(f"Sorry {name}, the number was {secret}.")
````

````python
# exercise 2

# Get user input
distance = float(input("Enter distance in km: "))
consumption = float(input("Enter fuel consumption (liters per 100km): "))
price_per_liter = float(input("Enter fuel price per liter: "))
passengers = float(input("Enter number of passengers: "))

# Calculate fuel needed
fuel_needed = (distance / 100) * consumption

# Calculate total fuel cost
total_fuel_cost = fuel_needed * price_per_liter

# Calculate cost per person
cost_per_person = total_fuel_cost / passengers

# Print the report
print("-------- Trip Cost Estimator --------")
print(f"Distance         : {distance} km")
print(f"Fuel needed      : {fuel_needed:.2f} L")
print(f"Total fuel cost  : {total_fuel_cost:.2f} DA")
print(f"Cost per person  : {cost_per_person:.2f} DA")
````

````python
# exercise 3

correct_username = "admin"
correct_password = "1234"

attempts = 3
while True:
    try:
        if attempts > 0:
            print("--- Login ---")
            username = input("Username: ")
            password = input("Password: ")

            if username == correct_username and password == correct_password:
                print(f"✅ Access granted. Welcome, {correct_username}!")
                print(f"username: {username}, password: {password}")
                break
            elif username == correct_username:
                attempts -= 1
                print(f"❌ Wrong password. Attempts left: {attempts}")
                print(f"username: {username}, password: {password}")
            else:
                attempts -= 1
                print(f"❌ Unknown user. Attempts left: {attempts}")
                print(f"username: {username}, password: {password}")
        else:
            print("🔒 Account locked.")
            break
    except Exception:
        print("Unexpected error.")
        break
````

````python
# exercise 4
import math

# Function 1 — circle_area
def circle_area(radius):
    return round(math.pi * math.pow(radius, 2), 2)

# Function 2 — rectangle_area
def rectangle_area(width, height):
    return width * height

# Function 3 — classify_shape
def classify_shape(area):
    if area < 50:
        return "Small"
    elif area < 200:
        return "Medium"
    elif area < 1000:
        return "Large"
    else:
        return "Huge"

# Function 4 — print_shape_report
def print_shape_report(shape_name, dimensions, area, size):
    print("-------- Shape Report --------")
    print(f"Shape      : {shape_name}")
    print(f"Dimensions : {dimensions}")
    print(f"Area       : {area:.2f}")
    print(f"Size       : {size}")

# Main program

shape = input("Enter shape (circle/rectangle): ").lower()
if shape == "circle":
    # Get radius from user
    radius = float(input("Enter radius: "))
    area = circle_area(radius)
    size = classify_shape(area)
    dimensions = f"radius = {radius}"
    print_shape_report("Circle", dimensions, area, size)

elif shape == "rectangle":
    # Get width and height from user
    width = float(input("Enter width: "))
    height = float(input("Enter height: "))
    area = rectangle_area(width, height)
    size = classify_shape(area)
    dimensions = f"width = {width}, height = {height}"
    print_shape_report("Rectangle", dimensions, area, size)

else:
    print("Unknown shape.")
````

````python
# exercise 5

# Part 1 — Data collection
scores = []
while True:
    user_input = input("Enter student score (or 'done' to finish): ")
    if user_input.lower() == "done":
        break
    if user_input.startswith("#"):
        continue
    try:
        score = float(user_input)
        if score < 0 or score > 100:
            print("Score out of range, skipping.")
            continue
        scores.append(score)
    except ValueError:
        print("Invalid score, skipping.")

# Part 2 — Analysis
total_students = len(scores)
sum_of_scores = 0
for score in scores:
    sum_of_scores += score
average = sum_of_scores / total_students if total_students > 0 else 0
# Search for passed and failed
passed = []
for score in scores:
    if score >= 60:
        passed.append(score)
# Search for failed
failed = []
for score in scores:
    if score < 60:
        failed.append(score)

# Search for 100
found = False
for score in scores:
    if score == 100:
        found = True
        break

# Find highest and lowest
highest = None
lowest = None
for score in scores:
    if highest is None or score > highest:
        highest = score
    if lowest is None or score < lowest:
        lowest = score

# Print the report
print("-------- Grade Report --------")
print(f"Total students : {total_students}")
print(f"Sum of scores  : {sum_of_scores}")
print(f"Class average  : {average:.1f}")
print(f"Passed (>= 60) : {', '.join(str(s) for s in passed)}")
print(f"Failed  (< 60) : {', '.join(str(s) for s in failed)}")
print(f"Searched 100   : {'Found' if found else 'Not found'}")
print(f"Highest score  : {highest}")
print(f"Lowest score   : {lowest}")

````
