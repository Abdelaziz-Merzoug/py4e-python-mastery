## 🎯 Chapter 5 Boss Exercise — "The Inventory Manager"

**What it tests:** while loop, for loop, break, continue, all 5 loop idioms, None pattern, is operator, filtering, list of numbers.

Write a program in two parts:

**Part 1 — Data collection (while loop):**
- Keep asking: `"Enter item price (or 'done' to finish): "`
- Use `continue` to skip any input that starts with `#`
- Use `break` when the user types `done`
- Use `try/except` to reject non-numeric input with `"Invalid price, skipping."`
- After conversion, use `continue` to skip any price that is `<= 0` with `"Price must be positive, skipping."`
- Store valid prices in a list called `prices`

**Part 2 — Analysis (separate for loops):**

Run all analyses and print this report:
```
-------- Inventory Report --------
Total items       : 7
Total value       : 4355.00
Average price     : 622.14
Expensive (>500)  : 1200.00, 800.00, 750.00, 999.00
Cheap     (<=500) : 250.00, 199.00, 157.00
Search 999        : Found
Most expensive    : 1200.00
Cheapest          : 157.00
```

**Constraints:**
- Part 1 must use `while True` with `break` and `continue` used **twice** (for `#` and for `<= 0`)
- Part 2 must use **completely separate `for` loops** for: count, sum, expensive filter, cheap filter, boolean search, most expensive, cheapest
- Average must be calculated as `total_value / total_items` after the loops — guard against empty list
- Most expensive and cheapest must use `None` + `is` — no `max()` or `min()`
- Boolean search must look for the price `999` and start with `found = False`
- All prices in the report must display with `:.2f`
- Total items must use a counting loop — no `len()`

````python
# exercise 1
while True:
    try:
        num = int(input("Enter a number between 1 and 10: "))
        if num < 1 or num > 10:
            print("Error: please enter a number between 1 and 10.")
            break
        print(f"---- Multiplication Table for {num} ----")
        for i in range(1, 11):
            result = num * i
            print(f"{num}  x  {i}  =  {result}")

        again = input("Do you want another table? (yes/no): ")
        if again.lower() != "yes":
            print("Goodbye!")
            break
    except ValueError:
        print("Error: please enter a valid number.")
````

````python
# exercise 2
# Exchange rates
dzd_to_usd = 0.0074
dzd_to_eur = 0.0068
usd_to_dzd = 135.14
usd_to_eur = 0.92
eur_to_dzd = 147.06
eur_to_usd = 1.09

# User input
try:
    amount = float(input("Enter amount: "))
    source = input("Enter source currency (DZD/USD/EUR): ")
    target = input("Enter target currency (DZD/USD/EUR): ")

    # Determine exchange rate
    if source == "DZD" and target == "USD":
        rate = dzd_to_usd
    elif source == "DZD" and target == "EUR":
        rate = dzd_to_eur
    elif source == "USD" and target == "DZD":
        rate = usd_to_dzd
    elif source == "USD" and target == "EUR":
        rate = usd_to_eur
    elif source == "EUR" and target == "DZD":
        rate = eur_to_dzd
    elif source == "EUR" and target == "USD":
        rate = eur_to_usd
    else:
        print("Unsupported conversion.")
        exit()

    # Perform conversion
    result = amount * rate

    # Print report
    print("-------- Currency Converter --------")
    print(f"Amount        : {amount:.2f} {source}")
    print(f"Converted to  : {result:.2f} {target}")
    print(f"Exchange rate : {rate:.4f}")

except ValueError:
    print("Error: please enter a valid number.")
````

````python
# Exercise 3
balance = 50000.0
pin = "4321"

# PIN check
attempts = 0
while attempts < 3:
    entered_pin = input("Enter your PIN: ")
    if entered_pin == pin:
        break
    else:
        attempts += 1

    if attempts == 3:
        print("Card blocked.")
        quit()

# ATM menu
while True:
    print("-------- ATM Menu --------")
    print("1. Check balance")
    print("2. Deposit")
    print("3. Withdraw")
    print("4. Exit")

    choice = input("Choose an option: ")
    if choice == "1":
        print(f"Your balance is: {balance:.2f} DZD")
    elif choice == "2":
        try:
            amount = float(input("Enter deposit amount: "))
            if amount <= 0:
                print("Invalid amount.")
                continue
            balance += amount
            print(f"Deposited. New balance: {balance:.2f} DZD")
        except ValueError:
            print("Invalid input.")
    elif choice == "3":
        try:
            amount = float(input("Enter withdrawal amount: "))
            if amount <= 0:
                print("Invalid amount.")
                continue
            if amount > balance:
                print("Insufficient funds.")
                continue
            balance -= amount
            print(f"Withdrawn. New balance: {balance:.2f} DZD")
        except ValueError:
            print("Invalid input.")
    elif choice == "4":
        print("Thank you. Goodbye!")
        break
    else:
        print("Invalid option.")
````

````python
# exercise 4

# Function definitions
def celsius_to_fahrenheit(c):
    return round((c * 9/5) + 32, 2)

def fahrenheit_to_celsius(f):
    return round((f - 32) * 5/9, 2)

def classify_temp(celsius):
    if celsius < 0:
        return "Freezing"
    elif 0 <= celsius <= 14:
        return "Cold"
    elif 15 <= celsius <= 24:
        return "Comfortable"
    elif 25 <= celsius <= 34:
        return "Warm"
    else:
        return "Hot"

def get_advice(label):
    if label == "Freezing":
        return "Wear a heavy coat!"
    elif label == "Cold":
        return "A jacket is recommended."
    elif label == "Comfortable":
        return "Enjoy the weather!"
    elif label == "Warm":
        return "Stay hydrated."
    elif label == "Hot":
        return "Avoid going out if possible."

def print_temp_report(original, unit, celsius, fahrenheit, label, advice):
    print("-------- Temperature Report --------")
    print(f"Input        : {original:.1f} {unit}")
    print(f"In Celsius   : {celsius:.1f}")
    print(f"In Fahrenheit: {fahrenheit:.1f}")
    print(f"Condition    : {label}")
    print(f"Advice       : {advice}")

# Main program

try:
    temp = float(input("Enter temperature: "))
    unit = input("Enter unit (C/F): ").upper()

    if unit == "C":
        celsius = temp
        fahrenheit = celsius_to_fahrenheit(celsius)
    elif unit == "F":
        fahrenheit = temp
        celsius = fahrenheit_to_celsius(fahrenheit)
    else:
        print("Unknown unit.")
        exit()

    label = classify_temp(celsius)
    advice = get_advice(label)
    print_temp_report(temp, unit, celsius, fahrenheit, label, advice)

except ValueError:
    print("Invalid input.")
````

````python
# exercise 5
# Part 1 - Data collection
prices = []
while True:
    entry = input("Enter item price (or 'done' to finish): ")
    if entry.lower() == "done":
        break
    if entry.startswith("#"):
        continue
    try:
        price = float(entry)
        if price <= 0:
            print("Price must be positive, skipping.")
            continue
        prices.append(price)
    except ValueError:
        print("Invalid price, skipping.")

# Part 2 - Analysis
# Total items
total_items = 0
for price in prices:
    total_items += 1

# total value
total_value = 0
for price in prices:
    total_value += price

# Average price
if total_items > 0:
    average_price = total_value / total_items
else:
    average_price = 0

# Expensive and cheap filters
expensive = []
cheap = []
for price in prices:
    if price > 500:
        expensive.append(price)
    else:
        cheap.append(price)

# Boolean search for 999
found_999 = False
for price in prices:
    if price == 999:
        found_999 = True
        break

# Most expensive and cheapest
most_expensive = None
cheapest = None
for price in prices:
    if most_expensive is None or price > most_expensive:
        most_expensive = price
    if cheapest is None or price < cheapest:
        cheapest = price

# Print report
print("-------- Inventory Report --------")
print(f"Total items       : {total_items}")
print(f"Total value       : {total_value:.2f}")
print(f"Average price     : {average_price:.2f}")
print(f"Expensive (>500)  : {', '.join(f'{p:.2f}' for p in expensive)}")
print(f"Cheap     (<=500) : {', '.join(f'{p:.2f}' for p in cheap)}")
print(f"Search 999        : {'Found' if found_999 else 'Not Found'}")
print(f"Most expensive    : {most_expensive:.2f}")
print(f"Cheapest          : {cheapest:.2f}")
````
