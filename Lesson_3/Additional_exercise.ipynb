### 🔀 Lesson 3: Conditional Execution & Error Handling

### Exercise 1: The Smart Thermostat
**Goal:** Write a program that asks the user for the current temperature in Celsius and classifies the weather. It must also handle bad input gracefully.

**Requirements:**
* Prompt the user for a temperature using `input()`.
* Use `try` and `except` to print `"Error: Please enter a valid number"` if the user types text instead of a number.
* If the input is valid, use `if/elif/else` to print:
  * "Freezing" if the temperature is below 0.
  * "Cold" if it is between 0 and 15 (inclusive).
  * "Warm" if it is between 16 and 30 (inclusive).
  * "Hot" if it is above 30.

### Exercise 2: The Bookstore Discount
**Goal:** Create a program that calculates a customer's final bill based on their purchase amount and membership status. This will test your ability to use `and` / `or` logic.

**Requirements:**
* Prompt the user for their total purchase amount (as a float).
* Prompt the user asking `"Are you a member? (yes/no)"`.
* Apply a **20% discount** IF the purchase is over $50 `and` they are a member.
* Apply a **10% discount** IF the purchase is over $50 `or` they are a member (but not both).
* Apply a **0% discount** otherwise.
* Print the final calculated price.


```python
# solution for Exercise 1

# prompt the user for temperature
temp_input = input("Please enter the current temperature in Celsius: ")

try:
    # try to convert the input to a float
    temperature = float(temp_input)

    # classify the weather based on the temperature
    if temperature < 0:
        print("Freezing")
    elif 0 <= temperature <= 15:
        print("Cold")
    elif 16 <= temperature <= 30:
        print("Warm")
    else:
        print("Hot")
except ValueError:
    # handle the case where the input is not a valid number
    print("Error: Please enter a valid number")
```

```python
# solution for Exercise 2

# prompt the user for total purchase amount
purchase_input = input("Please enter your total purchase amount: ")

try:
    # try to convert the input to a float
    purchase_amount = float(purchase_input)

    # prompt the user for membership status
    membership_input = input("Are you a member? (yes/no): ").strip().lower()

    # determine if the user is a member
    is_member = membership_input == "yes"

    # calculate the discount based on the conditions
    if purchase_amount > 50 and is_member:
        discount = 0.20  # 20% discount
    elif purchase_amount > 50 or is_member:
        discount = 0.10  # 10% discount
    else:
        discount = 0.00  # no discount

    # calculate the final price after applying the discount
    final_price = purchase_amount * (1 - discount)

    print(f"Your final price after discount is: ${final_price:.2f}")
except ValueError:
    # handle the case where the input is not a valid number
    print("Error: Please enter a valid number for the purchase amount")
```
